```python
import csv
import os
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException, WebDriverException

# Configuration de Selenium
def init_driver():
    chrome_options = Options()
    chrome_options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")
    driver = webdriver.Chrome(options=chrome_options)
    return driver

# Fonction de défilement
def scroll_and_wait(driver):
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5)  # Augmenter le temps pour donner plus de temps au chargement
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

# Fonction pour extraire les informations de chaque produit
def scrape_product_page(driver, url):
    driver.get(url)
    print("Chargement de l'URL:", url)

    # Attendre que le titre du produit soit visible
    WebDriverWait(driver, 30).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "h1.product-title"))
    )

    # Extraire le titre
    title = driver.find_element(By.CSS_SELECTOR, "h1.product-title span").text.strip()

    # Extraire la description
    description = driver.find_element(By.CSS_SELECTOR, "div#description-short p").text.strip()

    # Extraire les caractéristiques
    features = {}
    feature_rows = driver.find_elements(By.CSS_SELECTOR, "table.tableau_caracteristique tbody tr")
    for row in feature_rows:
        cells = row.find_elements(By.TAG_NAME, "td")
        if len(cells) == 2:
            features[cells[0].text.strip()] = cells[1].text.strip()

    # Extraire le prix
    price = driver.find_element(By.CSS_SELECTOR, "span.our_price_display").text.strip()

    # Extraire la saveur
    flavor = driver.find_element(By.CSS_SELECTOR, "span.feature-value").text.strip()

    return title, description, features, price, flavor

# Fonction principale pour lire les liens et scraper les informations
def scrape_website(csv_file, output_file):
    # Initialisation de Selenium
    driver = init_driver()

    # Vérifier et créer le fichier CSV de sortie s'il n'existe pas
    if not os.path.exists(output_file):
        with open(output_file, mode='w', newline='', encoding='utf-8') as file:
            fieldnames = ['lien_produit', 'nom_produit', 'Title', 'Description', 'Features', 'img_produit', 'prix_produit', 'saveur']
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            writer.writeheader()

    # Lecture des liens dans le fichier CSV
    links = []
    products_data = []
    try:
        with open(csv_file, mode='r', newline='', encoding='utf-8') as file:
            reader = csv.DictReader(file)
            for row in reader:
                links.append(row['url_produit'])  # Changer ici pour 'url_produit' si la colonne s'appelle ainsi
                products_data.append(row)  # Enregistrer toutes les données des produits
    except Exception as e:
        print(f"Erreur lors de la lecture du fichier CSV : {e}")
        return

    # Scraper chaque produit
    for link in links:
        print(f"Scraping du produit {link}")
        try:
            # Récupérer les informations du produit
            title, description, features, price, flavor = scrape_product_page(driver, link)

            # Trouver le nom du produit et l'image depuis la liste des produits
            product_data = next((product for product in products_data if product['url_produit'] == link), None)
            if product_data:
                product_name = product_data.get('nom_produit', 'N/A')
                img_product = product_data.get('img_produit', 'N/A')
            else:
                product_name = 'N/A'
                img_product = 'N/A'

            # Sauvegarder dans le fichier CSV
            with open(output_file, mode='a', newline='', encoding='utf-8') as file:
                writer = csv.DictWriter(file, fieldnames=['lien_produit', 'nom_produit', 'Title', 'Description', 'Features', 'img_produit', 'prix_produit', 'saveur'])
                writer.writerow({
                    'lien_produit': link,
                    'nom_produit': product_name,
                    'Title': title,
                    'Description': description,
                    'Features': str(features),
                    'img_produit': img_product,
                    'prix_produit': price,
                    'saveur': flavor
                })
            print(f"Produit {product_name} scrappé avec succès.")

        except Exception as e:
            print(f"Erreur lors du scraping du produit {link}: {e}")

    driver.quit()

if __name__ == "__main__":
    # Fichier d'entrée avec les liens des produits
    csv_file = r"C:\Users\antoa\Desktop\projet certif\scrap_final\links\link_produits_boissons_ptitvap_copy.csv"
    # Fichier de sortie pour les données scrappées
    scrap_folder = 'donnees_scrappes'
    os.makedirs(scrap_folder, exist_ok=True)
    output_file = os.path.join(scrap_folder, 'produits_boisson_ptitvap.csv')

    # Scraper les produits
    scrape_website(csv_file, output_file)
```

```python
import csv
import re
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException, WebDriverException

def scroll_and_wait(driver):
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5)  # Augmenter le temps pour donner plus de temps au chargement
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

def scrape_website(url, output_filename):
    chrome_options = Options()
    chrome_options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")

    driver = webdriver.Chrome(options=chrome_options)

    try:
        driver.get(url)

        # Afficher l'URL actuelle et le titre de la page pour débogage
        print("URL actuelle:", driver.current_url)
        print("Titre de la page:", driver.title)

        # Attendre que le bandeau de cookies soit présent et accepter les cookies
        print("Attente du bandeau de cookies...")
        WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((By.ID, "cookie-banner-message"))
        )
        print("Le bandeau de cookies est présent.")

        # Faire défiler jusqu'au bouton d'acceptation des cookies
        accept_cookies_button = driver.find_element(By.CSS_SELECTOR, '#cookie-banner-message button.btn-vert')
        driver.execute_script("arguments[0].scrollIntoView(true);", accept_cookies_button)
        print("Accepter les cookies...")
        
        # Attendre que le bouton soit cliquable
        WebDriverWait(driver, 30).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, '#cookie-banner-message button.btn-vert'))
        )
        # Utiliser JavaScript pour cliquer sur le bouton si le clic classique échoue
        driver.execute_script("arguments[0].click();", accept_cookies_button)

        # Attendre que le formulaire de vérification de l'âge soit présent
        print("Attente du formulaire de vérification de l'âge...")
        WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((By.ID, "dateJour"))
        )
        print("Le formulaire de vérification de l'âge est présent.")

        # Remplir les champs jour, mois et année (remplacez par votre date de naissance)
        print("Remplissage des champs de date...")
        day_field = driver.find_element(By.ID, "dateJour")
        month_field = driver.find_element(By.ID, "dateMois")
        year_field = driver.find_element(By.ID, "dateAnnee")

        day_field.clear()
        day_field.send_keys("01")

        month_field.clear()
        month_field.send_keys("01")

        year_field.clear()
        year_field.send_keys("2000")

        # Attendre que les éléments de produit soient visibles et charger plus de produits
        print("Chargement des produits...")
        scroll_and_wait(driver)

        # Attendre que l'onglet "Avis" soit présent
        print("Attente de l'onglet 'Avis'...")
        WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "a.avisverifies_tab"))
        )
        print("Onglet 'Avis' trouvé.")

        # Cliquer sur l'onglet "Avis"
        avis_tab = driver.find_element(By.CSS_SELECTOR, "a.avisverifies_tab")
        driver.execute_script("arguments[0].click();", avis_tab)
        print("Clic sur l'onglet 'Avis'...")

        # Attendre que le contenu des avis se charge
        WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((By.ID, "ajax_comment_content"))
        )
        print("Contenu des avis chargé.")

        # Scroller pour charger tous les avis (si nécessaire)
        scroll_and_wait(driver)

        # Extraire les avis
        reviews = driver.find_elements(By.CSS_SELECTOR, "div.reviewAV")
        review_data = []

        for review in reviews:
            name = review.find_element(By.CSS_SELECTOR, "li.name").text.strip()
            date_review = review.find_element(By.CSS_SELECTOR, "li.dateReview b").text.strip()
            date_order = review.find_element(By.CSS_SELECTOR, "li.dateOrder b").text.strip()
            rating = review.find_element(By.CSS_SELECTOR, "li.rateAV b").text.strip()
            comment = review.find_element(By.CSS_SELECTOR, "div.review-item").text.strip()
            
            review_data.append([name, date_review, date_order, rating, comment])
            print(f"Nom: {name}, Date de l'avis: {date_review}, Date de commande: {date_order}, Note: {rating}, Commentaire: {comment}")

        # Sauvegarder les données dans un fichier CSV
        output_filename = 'avis_clients.csv'
        with open(output_filename, 'w', newline='', encoding='utf-8') as csvfile:
            csvwriter = csv.writer(csvfile)
            csvwriter.writerow(['Nom', 'Date de l\'avis', 'Date de commande', 'Note', 'Commentaire'])
            csvwriter.writerows(review_data)
            print(f"Les données des produits ont été sauvegardées dans '{output_filename}'.")

    except NoSuchElementException as e:
        print(f"Élément non trouvé : {str(e)}")
    except TimeoutException as e:
        print(f"Dépassement de délai lors de l'attente d'un élément : {str(e)}")
    except WebDriverException as e:
        print(f"Erreur WebDriver : {str(e)}")
    except Exception as e:
        print(f"Une erreur inattendue est survenue : {str(e)}")

    finally:
        driver.quit()

def get_filename_from_url(url):
    # Utiliser une expression régulière pour extraire le dernier mot après le dernier tiret
    match = re.search(r'\/([^\/]+)$', url)
    if match:
        last_part = match.group(1).split('-')[-1]
        return f"{last_part}_links.csv"
    else:
        raise ValueError("Le format de l'URL ne correspond pas aux attentes.")

if __name__ == "__main__":
    urls = [
        "https://www.lepetitvapoteur.com/fr/le-petit-vapoteur/5663-e-liquide-peche-le-petit-vapoteur.html"
    ]
    
    for url in urls:
        # Générer le nom de fichier basé sur le dernier mot de l'URL
        filename = get_filename_from_url(url)
        scrape_website(url, filename)
```
